```python
def distribute_ratings(formation, total_rating):
    # Split the formation string into a list of numbers
    positions = formation.split('-')

    # Create a list to store the number of players in each position
    num_players = [int(p) for p in positions]

    # Calculate the total number of players
    total_players = sum(num_players)

    # Calculate the percentage of the total rating for each position
    ratings = [total_rating * (num / total_players) for num in num_players]

    return ratings

position_ratings = distribute_ratings('3-5-2', 760)
```

```python
# Iterate over each row in sim_data
for index, row in df1.iterrows():
    # Extract the formation and ratings for home team
    home_formation = row['formation']
    home_defense = row['defence rating']
    home_midfield = row['midfield rating']
    home_attack = row['attack rating']
    total_home_rating = home_defense + home_midfield + home_attack

    # Calculate the distributed ratings for home team
    home_ratings = distribute_ratings(home_formation, int(total_home_rating))
    home_defense_rating = home_ratings[0]
    home_midfield_rating = sum(home_ratings[1:-1])
    home_attack_rating = home_ratings[-1]

    # Extract the formation and ratings for away team
    away_formation = row['away formation']
    away_defense = row['away defence rating']
    away_midfield = row['away midfield rating']
    away_attack = row['away attack rating']
    total_away_rating = away_defense + away_midfield + away_attack

    # Calculate the distributed ratings for away team
    away_ratings = distribute_ratings(away_formation, total_away_rating)
    away_defense_rating = away_ratings[0]
    away_midfield_rating = sum(away_ratings[1:-1])
    away_attack_rating = away_ratings[-1]

    # Update the sim_data DataFrame with the redistributed ratings
    df1.at[index, 'defence rating'] = int(home_defense_rating)
    df1.at[index, 'midfield rating'] = int(home_midfield_rating)
    df1.at[index, 'attack rating'] = int(home_attack_rating)
    df1.at[index, 'away defence rating'] = int(away_defense_rating)
    df1.at[index, 'away midfield rating'] = int(away_midfield_rating)
    df1.at[index, 'away attack rating'] = int(away_attack_rating)
```
